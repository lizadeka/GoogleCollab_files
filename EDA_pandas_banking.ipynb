```python
d = {"no": 0, "yes": 1}
df["y"] = df["y"].map(d)
df.head()
```

Indexing and Extracting data

```python
print("Share of attracted clients =", '{:.1%}'.format(df["y"].mean()))
```

```python
# df[df["y"] == 1].mean()
```

```python
df[df["y"] == 1].mean(numeric_only=True)
```

```python
acd = round(df[df["y"] == 1]["duration"].mean(), 2)
acd_in_min = acd // 60
print("Average call duration for attracted clients =", acd_in_min, "min", int(acd) % 60, "sec")
```

```python
print("Average age of attracted clients =", int(df[(df["y"] == 1) & (df["marital"] == "single")]["age"].mean()), "years")
# The average age of unmarried attracted clients is 31, which should be considered when working with such clients.
```

```python
df[-1:]
```

Pivot tables

```python
pd.crosstab(df["y"], df["marital"])
```

```python
pd.crosstab(df["y"],
            df["marital"],
            normalize = 'index')
# more than half of the clients (61%, column married) are married and have not issued a deposit.
```

```python
df.pivot_table(
    ["age", "duration"],
    ["job"],
    aggfunc = "mean",
).head(10)
# average age and the call duration for different types of client employment job
```

Visualization in Pandas

```python
pd.plotting.scatter_matrix(
    df[["age", "duration", "campaign"]],
    figsize = (15, 15),
    diagonal = "kde")
plt.show()
```

```python
df["age"].hist()
# most of our clients are between the ages of 25 and 50, which corresponds to the actively working part of the population.
```

```python
df["duration"].hist()
```

```python
df["campaign"].hist()
```

```python
df.hist(color = "k",
        bins = 30,
        figsize = (15, 10))
plt.show()
```

```python
df.boxplot(column = "age",
           by = "marital")
plt.show()
# unmarried people are on average younger than divorced and married ones. For the last two groups, there is an outlier zone over 70 years old, and for unmarried - over 50.
```

```python
df.boxplot(column = "age",
           by = ["marital", "housing"],
           figsize = (20, 20))
plt.show()
# age and marital status do not have any significant influence on having a housing loan.
```

Tasks

```python
# List 10 clients with the largest number of contacts.
df.sort_values(by = "campaign", ascending = False).head(10)
```

```python
# Determine the median age and the number of contacts for different levels of client education.
df.pivot_table(
    ["age", "campaign"],
    ["education"],
    aggfunc = ["mean", "count"],
)
```
